```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv('anggur.csv')
sns.set_theme()
```

```python
# Histogram
for column in df.columns:
    # separate each graph
    plt.figure()
    # plot histogram for current column
    sns.histplot(df, x = column)
    
```

```python
# BoxPlot
for column in df.columns:
    # separate each graph
    plt.figure()
    # plot histogram for current column
    sns.boxplot(df, x = column)
```
