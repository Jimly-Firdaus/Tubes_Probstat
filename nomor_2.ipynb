```python
# Setup data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('anggur.csv')
sns.set_theme()
```

# 2. Visualisasi Data untuk setiap kolom
### 2.1 Histogram Plot untuk setiap kolom secara berurutan

```python
n_cols = len(df.columns)
fig, axes = plt.subplots(n_cols, 2, figsize=(10, 5*n_cols))
for i, column in enumerate(df.columns):
    sns.histplot(df[column], ax=axes[i, 0])
    axes[i, 0].set_title(column)
    axes[i, 0].set_xlabel(column)
    axes[i, 0].set_ylabel('Frekuensi')
    sns.boxplot(y=df[column], ax=axes[i, 1])
    axes[i, 1].set_title(column)
    axes[i, 1].set_xlabel(column)
    axes[i, 1].set_ylabel('Nilai')
plt.tight_layout()
plt.show()
```
