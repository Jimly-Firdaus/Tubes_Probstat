```python
# Setup data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Load data
data = pd.read_csv('anggur.csv')
```

<h3 align="center"><u>Shapiro-Wilk Test</u></h3>

Shapiro-Wilk test adalah sebuah metode statistik yang digunakan untuk menguji apakah sampel data berasal dari populasi yang berdistribusi normal. Tes ini dinamai sesuai dengan Samuel Shapiro dan Martin Wilk, yang mengembangkan tes ini pada tahun 1965.

Dalam tes Shapiro-Wilk, hipotesis nol (H0) menyatakan bahwa data berasal dari populasi yang berdistribusi normal. Hipotesis alternatif (H1) menyatakan bahwa data tidak berasal dari populasi yang berdistribusi normal.

Untuk melakukan tes Shapiro-Wilk, kita menghitung nilai statistik W dan nilai p-value yang sesuai. Nilai W adalah ukuran seberapa baik data cocok dengan distribusi normal. Nilai p-value menunjukkan probabilitas mendapatkan nilai W yang sama atau lebih ekstrem jika data benar-benar berasal dari populasi yang berdistribusi normal.

<h4 align="center">W = (b’x)² / S²</h4>
Di mana:

* x adalah vektor data yang diurutkan dalam urutan menaik
* b adalah vektor konstanta yang dihitung dari matriks kovariansi dari urutan statistik normal
* S² adalah varians sampel

Setelah menghitung nilai W dan p-value, kita dapat menggunakan nilai p-value untuk membuat keputusan tentang hipotesis nol. Jika p-value lebih besar dari tingkat signifikansi yang telah ditentukan (misalnya, 0,05), kita gagal menolak hipotesis nol dan menyimpulkan bahwa data berasal dari populasi yang berdistribusi normal. Jika p-value lebih kecil dari tingkat signifikansi yang telah ditentukan, kita menolak hipotesis nol dan menyimpulkan bahwa data tidak berasal dari populasi yang berdistribusi normal.

```python
# Loop through each column in data
for col in data.columns:
    # check if column numeric
    if np.issubdtype(data[col].dtype, np.number):
        # plot histogram
        sns.histplot(data=data, x=col, kde=True)
        plt.title(col)
        plt.show()
        
        # perform Shapiro-Wilk test
        stat, p = shapiro(data[col])
        print('Shapiro-Wilk test for', col)
        print('Nilai statistik (W) = %.3f, p-value = %.3f' % (stat, p))
        
        # results
        alpha = 0.05
        if p > alpha:
            print(col, 'berdistribusi normal (gagal menolak H0)')
        else:
            print(col, 'tidak berdistribusi normal (tolak H0)')
```

To determine whether a histogram looks Gaussian or not, you can use the Shapiro-Wilk test. The Shapiro-Wilk test tests the null hypothesis that a sample comes from a normal distribution. The output of the test will include a p-value. If the p-value is greater than the significance level (usually 0.05), then we fail to reject the null hypothesis and conclude that the sample comes from a normal distribution. If the p-value is less than or equal to the significance level, then we reject the null hypothesis and conclude that the sample does not come from a normal distribution.

In general, if a histogram looks like a bell curve (i.e., symmetric and unimodal), then it is likely to be normally distributed. However, this is not always the case, especially for small sample sizes. That’s why it’s important to use a normality test like the Shapiro-Wilk test to confirm whether a sample is normally distributed or not.
