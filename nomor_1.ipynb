### Data Modus untuk setiap kolom berurutan

```python
# Modus
for column in df.columns:
    # get mode for current col
    modes = df[column].mode()
    print(column)
    print("Total Modes: " + str(len(modes)))
    # print all possible mode
    for i in range (len(modes) - 1):
        print(str(modes[i]) + ", ", end="")
    print(str(modes[len(modes) - 1]))
    print("--------------\n")

```

### Data Variansi untuk setiap kolom berurutan

```python
# Variansi
for column in df.columns:
    # convert column to numerical data
    column_data = pd.to_numeric(df[column], errors='coerce')

    # drop rows with missing values
    column_data = column_data.dropna()

    # get variance for current column
    var = column_data.var()

    print(column)
    print("Variance : " + str(var))
    print("--------------\n")

```

### Data Range untuk setiap kolom berurutan

```python
# Range
for column in df.columns:
    # convert column to numerical data
    column_data = pd.to_numeric(df[column], errors='coerce')
    max_column_data = column_data.max()
    min_column_data = column_data.min()

    # get range for current column
    range = float(max_column_data) - float(min_column_data)
    print(column + "\n")
    print("Range : " + str(range))
    print("\n--------------")
```

### Data IQR untuk setiap kolom secara berurutan

```python
# IQR
for column in df.columns:
    # convert column to numerical data
    column_data = pd.to_numeric(df[column], errors='coerce')
    iqr = column_data.quantile(.75) - column_data.quantile(.25)
    # get range for current column
    
    print(column + "\n")
    print("IQR : " + str(iqr))
    print("\n--------------")
```

### Data Skewness untuk setiap kolom secara berurutan

```python
# Skewness
for column in df.columns:
    # convert column to numerical data
    column_data = pd.to_numeric(df[column], errors='coerce')
    skew = column_data.skew()
    # get range for current column
    
    print(column + "\n")
    print("Skew : " + str(skew))
    print("\n--------------")
```

### Data Kurtosis untuk setiap kolom secara berurutan

```python
# Kurtosis
for column in df.columns:
    # convert column to numerical data
    column_data = pd.to_numeric(df[column], errors='coerce')
    kurtosis = column_data.kurtosis()
    # get range for current column
    
    print(column + "\n")
    print("Kurtosis : " + str(kurtosis))
    print("\n--------------")
```
